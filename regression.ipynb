1. Import Required Packages

```python
import pandas as pd                                                       
import numpy as np                                                      
from xgboost import XGBRegressor                                                     
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier                                                     
from sklearn.svm import SVC     
from sklearn.preprocessing import StandardScaler                                                   
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, RandomForestClassifier, GradientBoostingClassifier                                                     
from sklearn.linear_model import LinearRegression  
from sklearn.model_selection import train_test_split                                                     
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, accuracy_score, confusion_matrix                                                      
print("All Packages Imported")
```

2. Read CSV Files

```python
df = pd.read_csv("wisconsin_breast_cancer.csv")
df
```

7. Check the Missing values in Percentage

```python
missing_percent = df.isnull().mean() * 100

print(missing_percent)
```

```python
df.isnull().sum()
```

8. Change the Columns Lower Case to  Upper Case.

```python
df.columns = df.columns.str.replace("_", " ").str.title()
```

9. Drop a Missing Columns and Values in Datasets.

```python
data = df.dropna()
```

10. Split X and Y.

```python
X = data.drop(['Id','Class'], axis=1)
y = data['Class']
```

11. Split Train, Test, Model Selectionand Train.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.3, random_state=42)
model = SVC()
model.fit(X_train, y_train)
```

12. Prediction and Check Model Accuracy.

```python
predict = model.predict(X_test)
print("R2 Score:", r2_score(y_test, predict))
print("Accuracy Score:", accuracy_score(y_test, predict))
print("MSE Error:", mean_squared_error(y_test, predict))
print("MBE Error:", mean_absolute_error(y_test, predict))
```

```python
import pandas as pd
df_compare = pd.DataFrame({
    'Actual': y_test.values,
    'Predicted': predict
})

df_compare.iloc[0:20]
```

# RandomForest Model.

1. Model Selection and Model Train, Predict.

```python
from sklearn.ensemble import RandomForestClassifier

random_model = RandomForestClassifier(n_estimators=200, max_depth=10, random_state=42)
random_model.fit(X_train, y_train)
y_random_pred = random_model.predict(X_test)
```

2. Model Accuracy.

```python
print("R2 Score:", r2_score(y_test, y_random_pred))
print("Accuracy Score:", accuracy_score(y_test, y_random_pred))
print("MSE Error:", mean_squared_error(y_test, y_random_pred))
print("MBE Error:", mean_absolute_error(y_test, y_random_pred))
```

```python
import pandas as pd
df_compare_random = pd.DataFrame({
    'Actual': y_test.values,
    'Predicted': y_random_pred
})

df_compare_random.iloc[10:30]
```

# DecisionTree Model.

1. Model Selection Traine, Predict.

```python
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

2. Model Accuracy.

```python
from sklearn.metrics import classification_report, confusion_matrix

print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

3. Compare Actual VS Predict.

```python
import pandas as pd
df_compare_xgboost = pd.DataFrame({
    'Actual': y_test.values,
    'Predicted': y_pred
})

df_compare_xgboost.iloc[10:30]
```
